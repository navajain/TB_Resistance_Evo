```python
base_directory = '/content/drive/MyDrive/EVO/rif_embeddings_v1/'
embed_dir = base_directory + 'embeds_1.0_singles/'

bad_samples = []
df = pd.read_csv(base_directory + 'rif_labels.csv')
for i, row in tqdm(df.iterrows()):
  if i < 6000: continue

  site = 'site_' + row['UNIQUEID'].split('.')[1] + '/'
  filename = embed_dir + site + row['UNIQUEID'] + '.npy'

  # does the embedding file exist?
  if not os.path.exists(filename):
    bad_samples.append(row['UNIQUEID'])
    continue

  # is the embedding file a reasonable size?
  # a (500, 4096) size embedding should be approx. 1.5 MB
  if os.path.getsize(filename) <= 1000000:
    bad_samples.append(row['UNIQUEID'])
    continue

  # is the shape of the embedding correct?
  emb = np.load(filename)
  if len(emb.shape) != 2 or emb.shape[0] != 500 or emb.shape[1] != 4096:
    bad_samples.append(row['UNIQUEID'])
```
